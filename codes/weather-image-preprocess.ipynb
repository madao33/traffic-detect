## 预处理Two-class Weather Image

将该数据集下的图片重命名为分类加序号并拷贝到MWD对应分类目录下

```python
import os
import cv2
save_path ='../datasets/'
```

```python
tw_path1 = '../datasets/Two-class Weather Image/train'
tw_path2 = '../datasets/Two-class Weather Image/test'
```

```python
def save2path(tw_path, save_path, label):
    folderlist = os.listdir(tw_path)
    for folder in folderlist:
        inner_path = os.path.join(tw_path, folder)
        filelist = os.listdir(inner_path)
        total_num_file = len(filelist)
        cnt = 0
        for item in filelist:
            src = os.path.join(inner_path, item)
            dst = os.path.join(save_path, folder) + '/' + label + '_' + str(cnt)+'.jpg'
            try:
                os.rename(src, dst)
                cnt += 1
                if cnt*10 / total_num_file == 1: 
                    print("converting %d\%" %(cnt / total_num_file))
            except:
                continue
```

```python
save2path(tw_path1, save_path, '')
```

```python
save2path(tw_path2, save_path, 'tc')
```

```python
MCI_path = '../datasets/Multi-class Weather Image/classification/weather_classification'
save2path(MCI_path, save_path, 'mci')
```

```python
MCD_path = '../datasets/Multi-class Weather Dataset'
save2path(MCD_path, save_path, 'mcd')
```
